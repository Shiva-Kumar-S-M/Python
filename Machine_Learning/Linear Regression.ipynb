```python
#Importing the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#generating datasets
np.random.seed(42)
X=np.random.rand(5000,1)*100
Y=4*X+np.random.randn(5000,1)*20
```

```python
#creating and training Linear Regression 
model=LinearRegression()
model.fit(X,Y)
```

```python
#predicting values
Y_pred=model.predict(X)
```

```python
#Visualizing the regression line
plt.figure(figsize=(6,4))
plt.scatter(X,Y,label='Data points',color='Green')
plt.plot(X,Y_pred,color='Red',label='Regression Line',linewidth=5)
plt.title("Linear regression with random dataset")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#slope and intercept
print("Slope(coefficiant)=",model.coef_[0][0])
print("Intercept=",model.intercept_[0])
```
