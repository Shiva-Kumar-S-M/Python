```python
#Generating datasets for binary classifications

from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X,y=make_circles(n_samples=400,shuffle=True,noise=0.1,random_state=58)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show() 
```

```python
#to generate 2d classification dataset
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

x,y=make_classification(n_samples=1000,
                        n_features=2,
                        n_redundant=0,
                        n_informative=2,
                        n_classes=3,
                        n_repeated=0,
                        n_clusters_per_class=1)
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
from sklearn.datasets import make_multilabel_classification
import pandas as pd
import matplotlib.pyplot as plt

x,y=make_multilabel_classification(n_samples=2000,n_features=2,
                                    n_classes=2,n_labels=2,random_state=23)
df=pd.concat([pd.DataFrame(x,columns=["X1","X2"]),
              pd.DataFrame(y,columns=["Label1","Label2"])],
              axis=1)
display(df.head())
plt.scatter(df["X1"],df["X2"],c=df["Label1"])
plt.show()

```

```python
from sklearn.datasets import make_sparse_uncorrelated
import matplotlib.pyplot as plt

x,y=make_sparse_uncorrelated(n_samples=1000,n_features=4,random_state=23)
plt.figure(figsize=(12,10))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.scatter(x[:,i],y)
    plt.xlabel('x'+str(i+1))
    plt.ylabel('y')
plt.show()
```

```python
from sklearn.datasets import make_friedman2
import matplotlib.pyplot as plt

x,y=make_friedman2(n_samples=1000,random_state=23)
plt.figure(figsize=(12,10))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.scatter(x[:,i],y)
    plt.xlabel('x'+str(i+1))
    plt.ylabel('y')
plt.show()
```
