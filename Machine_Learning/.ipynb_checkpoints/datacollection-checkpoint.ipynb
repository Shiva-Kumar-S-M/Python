```python
#Generating datasets for binary classifications

from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X,y=make_circles(n_samples=400,shuffle=True,noise=0.1,random_state=58)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show() 
```

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X,y=make_circles(n_samples=3000,shuffle=True,noise=0.01,random_state=42)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
#Example 2: Two interlocking half circles represent the 2d binary classification data produced by the make_moons() function.

from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X,y=make_moons(n_samples=500,shuffle=False,noise=0.1,random_state=48)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X,y=make_moons(n_samples=500,shuffle=True,noise=0.1,random_state=48)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X,y=make_blobs(n_samples=500,centers=3,n_features=2,random_state=48)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X,y=make_blobs(n_samples=10000,centers=4,n_features=2,random_state=100)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
#Generating datasets for linear regression 
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

X,y=make_regression(n_samples=1000,n_features=1,noise=15,random_state=23)
plt.scatter(X,y)
plt.show()
```
